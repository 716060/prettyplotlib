## `prettyplotlib.scatter` Example

The default `matplotlib` color cycle is not pretty to look at ... It was taken from MATLAB's color cycle. 

Need to do the `matplotlib` example first because `prettyplotlib` changes the default color cycle to a nicer one, from [ColorBrewer](http://bl.ocks.org/mbostock/5577023)'s Set2. What's even worse is that if you just do a `scatter` plot, then it doesn't cycle at all...

```python
import matplotlib.pyplot as mpl_plt
# Set the random seed for consistency
np.random.seed(12)

fig, ax = mpl_plt.subplots(1)

# Show the whole color range
for i in range(8):
    x = np.random.normal(loc=i, size=1000)
    y = np.random.normal(loc=i, size=1000)
    ax.scatter(x, y, label=str(i))
ax.legend()
    
ax.set_title('prettyplotlib `scatter` example\nshowing default matplotlib `scatter`')
fig.savefig('scatter_matplotlib_default.png')
```

```python
import prettyplotlib as ppl
from prettyplotlib import plt
from prettyplotlib import mpl
from prettyplotlib import brewer2mpl

# Set the random seed for consistency
np.random.seed(12)

fig, ax = plt.subplots(1)

# Show the whole color range
for i in range(8):
    x = np.random.normal(loc=i, size=1000)
    y = np.random.normal(loc=i, size=1000)
    ppl.scatter(ax, x, y, label=str(i), facecolor='none')
    
ppl.legend(ax)
    
ax.set_title('prettyplotlib `scatter` example\nshowing default color cycle and scatter params')
fig.savefig('scatter_prettyplotlib_default.png')
```

Now that's pretty nice!

A few things happened here:

1. Default color cycle changed to the colors of ColorBrewer's Set2
2. Default scatter params are empty circle, with color on the circle line, to allow for better visualization of overlaps
3. Removed top and right axis lines
4. Ever-so-subtly changed the ticklabel and axis line colors from black (`#000000`) to almost black (`#262626`)
5. On the legend, removed the black frame and filled the body of the legend with a light grey. Also changed the number of points shown from three to one, since the user isn't dumb
